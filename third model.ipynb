```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Train.csv')
df.head()
```

```python
df.drop('Applicant_ID',axis=1,inplace=True)
```

```python
target = pd.get_dummies(df['default_status'],drop_first=True)
productType = pd.get_dummies(df['form_field47'],drop_first=True)
df = pd.concat([df,target,productType],axis=1)
df.drop(['default_status','form_field47'],axis=1,inplace=True)
df.head()
```

```python
df.fillna(-999999,inplace=True,axis=1)
```

```python
df.isnull().sum()
```

```python
X = df.drop(['yes'],axis=1)
y= df['yes']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion='gini', max_depth=7, n_estimators=200, random_state =0).fit(x_train,y_train)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
y_pred = forest.predict_proba(x_test)[:,1]
y_pred
```

```python
roc_auc_score(y_test,y_pred)
```

```python
classifier = RandomForestClassifier(n_jobs=-1)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid ={
    "n_estimators": [50,100,200,300,400],
    "max_depth": [1,3,5,7],
    "criterion": ["gini", "entropy"],
}

model = GridSearchCV(
    estimator=classifier,
    param_grid=param_grid,
    scoring='roc_auc',
    verbose=10,
    n_jobs=1,
    cv=25
)
```

```python
model.fit(X,y)
```

```python
model.best_score_
```

```python
model.best_estimator_.get_params
```

```python
from sklearn.metrics import confusion_matrix
from catboost import CatBoostClassifier
cat = CatBoostClassifier(learning_rate=0.03,
                           eval_metric='AUC',verbose=True)
cat.fit(x_train,y_train)
cat_pred = cat.predict(x_test)
```

```python
cat_pred = cat.predict_proba(x_test)[:,1]
```

```python
roc_auc_score(y_test,cat_pred)
```
